```python
import pandas as pd
```

```python
hh_df = pd.read_csv('../output/hanja_hangul_pairs.csv')
```

```python
hh_df['hanja_num_chars'] = hh_df['hanja_word'].apply(lambda x: len(x))
```

```python
hh_df['hangul_num_chars'] = hh_df['hangul_word'].apply(lambda x: len(x))
```

```python
hh_df
```

```python
hh_df['hangul_num_equal_hanja_num'] = hh_df.apply(
    lambda row: row['hangul_num_chars'] == row['hanja_num_chars'],axis=1)
```

```python
hh_df[hh_df['hangul_num_equal_hanja_num'] == False]
```

```python
hanja_hangul_pairs = [
    (hanja_char, hangul_char)
    for row_dict in hh_df.to_dict(orient='records')
    for hanja_char, hangul_char in zip(
            list(row_dict['hanja_word']),
            list(row_dict['hangul_word']))
]
```

```python
hanja_hangul_pairs.sort()
```

```python
hanja_hangul_pairs_set = set(hanja_hangul_pairs)
```

```python
hanja_hangul_pairs_set
```

```python
hanja_hangul_chars_df = pd.DataFrame(hanja_hangul_pairs_set).rename(
    columns={
        0: 'hanja_char',
        1: 'hangul_char'
    }
)
```

```python
repeat_hanja = hanja_hangul_chars_df.groupby('hanja_char').agg([list, 'count'])
```
